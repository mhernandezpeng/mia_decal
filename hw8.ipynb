```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
def make_sine_wave(x, A, w):
    return A*sin(w*x)
```

```python
x = np.linspace(0, 2*np.pi, 1000)
amplitudes = np.array([0.5, 1, 1.5, 2, 2.5])
frequencies = np.array([1, 2, 3, 4, 5])
```

```python
plt.figure(figsize = (5,5))
for amp, freq in zip(amplitudes, frequencies):
    y = amp * np.sin(freq * x)  
    plt.plot(x, y, label=f'A={amp}, f={freq}')  
plt.title('Sine waves with varied amplitudes and frequencies')
plt.xlabel('x (radians)')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```
